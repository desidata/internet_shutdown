```python
map_data=gpd.read_file(r"/Users/debodeepbanerjee/Library/CloudStorage/OneDrive-Personal/Documents/desidata/India States/Indian_states.shp")
map_data.head()
```

```python
merged = map_data.set_index('st_nm').join(df.set_index('State'))#.reset_index(level=0, inplace=True)
merged#.head()
```

```python
from matplotlib.colors import ListedColormap, Normalize, LogNorm
N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.linspace(177/256, 1, N)  # Red channel values (constant at 0)
vals[:, 1] = np.linspace(99/256, 1, N)  # Green channel values (constant at 0)
vals[:, 2] = np.linspace(255/256, 1, N)  # Blue channel values (from blue to violet)

# Invert the colormap
blue_to_violet_cmap =  ListedColormap(vals[::-1])

# Assuming 'merged' is a GeoDataFrame with a 'Percentage with iodised salt' column for plotting
fig = plt.figure(figsize=(12, 8), dpi=1200)
# Adjust normalization for the colormap
norm = plt.Normalize(vmin=merged['stats'].min(), vmax=merged['stats'].max())
# Customize the colorbar legend with reduced length (shrink=0.7)
sm = plt.cm.ScalarMappable(cmap=blue_to_violet_cmap,
                           norm=norm)
# Create the plot using the custom colormap, disable the GeoDataFrame legend
ax = merged.plot(column='stats', 
                 categorical=False, 
                 legend=False,
                 cmap=blue_to_violet_cmap, 
                 edgecolor='black', 
                 linewidth=0.25,
                missing_kwds={                 # Handle NaN (missing data)
                "color": "lightgrey",      # Color for NaN states
                "edgecolor": "black",      # Edge color for states
                "label": "No Data"})

sm._A = []
cbar = plt.colorbar(sm, ax=ax, pad=0.02, shrink=0.7)  # Adjust the shrink parameter as needed

# Remove x-axis and y-axis ticks and frame
plt.xticks([])
plt.yticks([])
ax.set_frame_on(False)

# Save the plot
plt.savefig('internet.svg', bbox_inches='tight')

# Show the plot (optional)
plt.show()
```
